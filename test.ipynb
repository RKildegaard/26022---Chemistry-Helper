```python
f = 2*x +3
a, b = 0, 5
n = 30
dx = (b - a) / n

# We want to approximate the integral of f = 2*x + 3 from a to b using rectangles.
# The interval [a, b] is split into n equal parts, each of width dx.

# Calculate left endpoint sum (rectangle method using left side of each interval)
left_sum = 0
for i in range(n):
    pt = a + i*dx
    left_sum += f.subs(x, pt)
left_sum *= dx

# Calculate right endpoint sum (rectangle method using right side of each interval)
right_sum = 0
for i in range(n):
    pt = a + (i+1)*dx
    right_sum += f.subs(x, pt)
right_sum *= dx

display(left_sum, right_sum)


""" g = x*y + y**2
integrate(g, (x, 0, 2), (y, 1, 3)) """
```
